```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([1, 2, 3, 4, 5])
Y = np.array([2, 4, 5, 4, 5])
```

```python
n = np.size(X)

m_x = np.mean(X)
m_y = np.mean(Y)

SS_xy = np.sum(Y * X) - n * m_y * m_x
SS_xx = np.sum(X * X) - n * m_x * m_x

b1 = SS_xy / SS_xx
b0 = m_y - b1 * m_x
```

```python
y_pred = b0 + b1 * X
```

```python
plt.scatter(X, Y, color = 'm', marker = 'o', s = 30)

plt.plot(X, y_pred, color = 'g')

plt.xlabel('X')
plt.ylabel('Y')

plt.show()
```
