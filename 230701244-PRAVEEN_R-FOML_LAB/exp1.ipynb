```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
```

```python

np.random.seed(0)
X_uni = 2 * np.random.rand(100, 1)
y_uni = 4 + 3 * X_uni + np.random.randn(100, 1)


X_train_uni, X_test_uni, y_train_uni, y_test_uni = train_test_split(X_uni, y_uni, test_size=0.2, random_state=42)


model_uni = LinearRegression()


model_uni.fit(X_train_uni, y_train_uni)


y_pred_uni = model_uni.predict(X_test_uni)


mse_uni = mean_squared_error(y_test_uni, y_pred_uni)
r2_uni = r2_score(y_test_uni, y_pred_uni)

print(f"Univariate Regression Model Coefficients: {model_uni.coef_[0][0]:.2f}")
print(f"Univariate Regression Model Intercept: {model_uni.intercept_[0]:.2f}")
print(f"Univariate Regression Mean Squared Error: {mse_uni:.2f}")
print(f"Univariate Regression R-squared: {r2_uni:.2f}")


plt.figure(figsize=(10, 6))
plt.scatter(X_test_uni, y_test_uni, color='blue', label='Actual data')
plt.plot(X_test_uni, y_pred_uni, color='red', linewidth=2, label='Regression line')
plt.xlabel('Independent Variable (X)')
plt.ylabel('Dependent Variable (y)')
plt.title('Univariate Linear Regression')
plt.legend()
plt.grid(True)
plt.show()
```

```python

np.random.seed(1)
X_bi_feat1 = 2 * np.random.rand(100, 1)
X_bi_feat2 = 3 * np.random.rand(100, 1)
X_bi = np.hstack((X_bi_feat1, X_bi_feat2))
y_bi = 5 + 2 * X_bi_feat1 + 1.5 * X_bi_feat2 + np.random.randn(100, 1) * 2


X_train_bi, X_test_bi, y_train_bi, y_test_bi = train_test_split(X_bi, y_bi, test_size=0.2, random_state=42)


model_bi = LinearRegression()
model_bi.fit(X_train_bi, y_train_bi)


y_pred_bi = model_bi.predict(X_test_bi)


mse_bi = mean_squared_error(y_test_bi, y_pred_bi)
r2_bi = r2_score(y_test_bi, y_pred_bi)

print(f"Bivariate Regression Model Coefficients: {model_bi.coef_[0]}")
print(f"Bivariate Regression Model Intercept: {model_bi.intercept_[0]:.2f}")
print(f"Bivariate Regression Mean Squared Error: {mse_bi:.2f}")
print(f"Bivariate Regression R-squared: {r2_bi:.2f}")


plt.figure(figsize=(10, 6))
plt.scatter(X_test_bi[:, 0], y_test_bi, c=X_test_bi[:, 1], cmap='viridis', label='Actual data (color by X2)')
plt.xlabel('Independent Variable 1 (X1)')
plt.ylabel('Dependent Variable (y)')
plt.title('Bivariate Linear Regression (X1 vs y, colored by X2)')
plt.colorbar(label='Independent Variable 2 (X2)')
plt.grid(True)
plt.show()
```

```python

np.random.seed(2)
X_multi_feat1 = 2 * np.random.rand(100, 1)
X_multi_feat2 = 3 * np.random.rand(100, 1)
X_multi_feat3 = 0.5 * np.random.rand(100, 1)
X_multi = np.hstack((X_multi_feat1, X_multi_feat2, X_multi_feat3))
y_multi = 10 + 2 * X_multi_feat1 - 1.5 * X_multi_feat2 + 4 * X_multi_feat3 + np.random.randn(100, 1) * 3


X_train_multi, X_test_multi, y_train_multi, y_test_multi = train_test_split(X_multi, y_multi, test_size=0.2, random_state=42)


model_multi = LinearRegression()
model_multi.fit(X_train_multi, y_train_multi)


y_pred_multi = model_multi.predict(X_test_multi)


mse_multi = mean_squared_error(y_test_multi, y_pred_multi)
r2_multi = r2_score(y_test_multi, y_pred_multi)

print(f"Multivariate Regression Model Coefficients: {model_multi.coef_[0]}")
print(f"Multivariate Regression Model Intercept: {model_multi.intercept_[0]:.2f}")
print(f"Multivariate Regression Mean Squared Error: {mse_multi:.2f}")
print(f"Multivariate Regression R-squared: {r2_multi:.2f}")


plt.figure(figsize=(10, 6))
plt.scatter(y_test_multi, y_pred_multi, alpha=0.7)
plt.plot([y_test_multi.min(), y_test_multi.max()], [y_test_multi.min(), y_test_multi.max()], 'r--', linewidth=2)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Multivariate Linear Regression: Actual vs Predicted')
plt.grid(True)
plt.show()
```
